<pre> <b>class sklearn.impute.SimpleImputer(missing_values=nan, strategy='mean'</b>, fill_value=None, verbose='deprecated', copy=True, add_indicator=False, keep_empty_features=False)

* <B>strategy 옵션</B>
    'mean'         : 평균값 (디폴트)

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')

df = pd.DataFrame(imp.fit_transform(df), columns=df.columns)
```

```python
df.isna().sum()
```

# 학습 & 평가

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score, precision_score, recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

## train test 분리

```python
y = df['glass_type']
X = df.drop(['glass_type'],  axis=1)
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=1111)
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

## 1차 점수 : 0.79070
* 결측 : 평균으로 채우기

```python
model = RandomForestClassifier(random_state=0)
model.fit(X_train, y_train)
pred = model.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(f"accuracy : {accuracy:.5f}")
```

## 피쳐 중요도 

```python
print( model.feature_importances_ )
plt.figure(figsize=(4,2))
s = pd.Series(model.feature_importances_, index=X_train.columns).sort_values()
sns.barplot(x=s.values, y=s.index)
plt.show()
```

# <font color=red><b>★★ 파생피쳐 </b></font>
* window_yn
<pre>
-- 1 building_windows_float_processed
-- 2 building_windows_non_float_processed
-- 3 vehicle_windows_float_processed
-- 4 vehicle_windows_non_float_processed (none in this database) XXXXX

-- 5 containers
-- 6 tableware
-- 7 headlamps

```python
df['windows_yn'] = df['glass_type'].replace([1,2,3,4,5,6,7],[1,1,1,1,0,0,0])
y = df['glass_type']
X = df.drop(['glass_type'],  axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=1111)
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

```python
high_corr_cols = ['al','ba','na','mg']

fig, axes = plt.subplots(1, 4,  figsize=(16, 4))
for i, col in enumerate(high_corr_cols):
    h = sns.distplot(df[df['windows_yn']==1][col], hist=True, kde=False, rug=True, bins=20, label="window_y", ax = axes[i])
    g = sns.distplot(df[df['windows_yn']==0][col], hist=True, kde=False, rug=True, bins=20, label="window_n"   , ax = axes[i])
    axes[i].legend()
    plt.xlabel(col)
plt.show()
plt.close()
```

## 2차 점수 : 0.88372
* 결측 : 평균으로 채우기
* 파생피쳐 : window_yn

```python
model = RandomForestClassifier(random_state=0)
model.fit(X_train, y_train)
pred = model.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(f"accuracy : {accuracy:.5f}")
```

## 피쳐 중요도 

```python
print( model.feature_importances_ )
plt.figure(figsize=(4,2))
s = pd.Series(model.feature_importances_, index=X_train.columns).sort_values()
sns.barplot(x=s.values, y=s.index)
plt.show()
```

# 교차 검증
* cross_val_score()

```python
from sklearn.model_selection import cross_val_score,  GridSearchCV, KFold, StratifiedKFold
```

## 3차 점수 : 0.85028 
* 결측 : 평균으로 채우기
* 파생피쳐 : window_yn
* 교차검증 : 5회 평균점수

```python
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
model = RandomForestClassifier(random_state=0) 

score_list = cross_val_score(model, X, y, scoring='accuracy' , cv=kf)
print( score_list )
print(f" {np.mean(score_list):.5f} " )
```

## 4차 점수 : 0.85504 
* 결측 : 평균으로 채우기
* 파생피쳐 : window_yn
* 교차검증 : 5회 평균점수

```python
mydic = {"n_estimators"      : [100,300],
         "max_depth"         : [10,15],
         "min_samples_split" : [1,2,3  ], 
         "min_samples_leaf"  : [1,2,3  ]
        }
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
models = GridSearchCV(model, scoring="accuracy", cv=kf ,  
                      param_grid = mydic)
models.fit(X, y)
print( models.best_params_ )
print(f" {models.best_score_:.5f} " )
```

## 직접 코딩 : 0.80411 

```python
from sklearn.model_selection import RandomizedSearchCV

model = RandomForestClassifier(bootstrap=True, ccp_alpha=0.0, class_weight=None,
                        criterion='gini', max_depth=15, max_features='auto',
                        max_leaf_nodes=None, max_samples=None,
                        min_impurity_decrease=0.0, min_impurity_split=None,
                        min_samples_leaf=1, min_samples_split=3,
                        min_weight_fraction_leaf=0.0, n_estimators=150,
                        n_jobs=-1, oob_score=False, random_state=11, verbose=0,
                        warm_start=False)

mydic = {"n_estimators"      : [100,150],
         "max_depth"         : [10,15,20],
         "min_samples_split" : [2,3,4 ], 
         "min_samples_leaf"  : [1,2  ]
        }

kf = StratifiedKFold(n_splits=10, random_state=None, shuffle=False)     

models = RandomizedSearchCV(model, scoring="accuracy", cv=kf ,  
                            param_distributions = mydic, random_state=22)

models.fit(X, y)
print( models.best_params_ )
print(f" {models.best_score_:.5f} " )
```
