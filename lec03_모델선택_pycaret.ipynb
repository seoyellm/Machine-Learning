<pre>
<font color=blue><b>데이터 나누기</b></font>
* train_size         : float, default = 0.7
* test_data          : pandas.DataFrame, default = None (train_size is ignored)
                       (test_data must be labeled and the shape match)
* data_split_shuffle : bool, default = True
* data_split_stratify: bool or list, default = False

<font color=blue><b>모델 검증</b></font>
* fold_strategy      : str or scikit-learn CV generator object, default = ‘stratifiedkfold’
                       ('kfold'  'stratifiedkfold'  'groupkfold'  'timeseries')
* fold               : int, default = 10
* fold_shuffle       : bool, default = False
* fold_groups        : str or array-like, with shape (n_samples,), default = None

<font color=blue><b>피쳐 가공</b></font>
* Missing Value Imputation
* One-Hot Encoding
* Train-Test Split

<font color=blue><b>피쳐선택</b></font>
* feature_selection           : bool, default = False  # 모든 피쳐 학습에 가담
* feature_selection_threshold : float, default = 0.8   # 피쳐의 80%만 학습에 가담

<font color=blue><b>다중공선</b></font>
* remove_multicollinearity    : bool, default = False  # True : 다중공선 삭제
* multicollinearity_threshold : float, default = 0.9   # 다중공선이 0.9 이상이면 삭제

<font color=blue><b>차원축소(PCA:Principal Component Analysis) </b></font>
* pca           : bool, default = False
* pca_method    : string, default = 'linear', 'kernel' , 'incremental'
* pca_components: int/float, default = 0.99  # 몇퍼센트의 설명력을 얻게 할 것인가?, 일반적으로 pc2~3 까지했을 때의 설명력(PC 몇개인지 찍어보기: pca.componets_)

<font color=blue><b>저분산 데이터 처리(Ignore Low Variance)</b>
    - Count of unique values in a feature  / sample size < 10%   # sample size가 10%이하면 버려라
    - Count of most common value / Count of second most common value > 20 times</font>
* ignore_low_variance: bool, default = False

<font color=blue><b>기타 </b></font>
session_id      : int, default = None (랜덤시드)
silent          : bool, default = False
profile         : bool, default = False  # 레포트 생성
profile_kwargs  : dict, default = {} (empty dict) # 레포트 생성시 속성 설정

n_jobs          : int, default      = -1
use_gpu         : bool or str, default = False
html            : bool, default = True  (IDE False / Jupyter True)  
verbose         : bool, default = True  # 로그메세지

custom_pipeline : (str, transformer) or list of (str, transformer), default = None  # 사용자가 만든 파이프라인 사용 여부
preprocess      : bool, default = True

## 앙상블
> <b>bagging</b>    
- ex. RandomForest
1. <font color='red'>모델 하나로 다양한 데이터를 학습</font>   --> 병렬    
2. 동시적 학습 가능(parallel) -> 속도 빠름
3. voting(투표)한 결과대로 결과 분류
    - 하드(Hard Voting) : 다수결
    - 소프트(Soft Voting) : 가중치 부여(조화 평균), 각 레이블 예측 확률의 평균으로 최종 분류(확률값을 w로 사용)  --> 일반적으로 머신러닝은 소프트 사용(특별한 경우 하드로 설정 가능)

> <b>boosting</b>
1. 가지치기 후 w(가중치)를 보정하여 다시 가지치기 --> overfitting의 문제
2. 다양한 트리 모델(트리 업데이트)로 데이터 하나를 학습
    - 학습이 잘 된 가지치기 쪽으로 <font color='red'>강한 w 부여하여 모델 업데이트</font>
3. overfittng 해결하기 위한 방법 : 모든 boosting 모델에는 <font color='red'>규제</font>가 들어있음 
4. 학습이 한번 끝날 때마다 가지치기의 모양이 달라짐
    - 순차적으로 움직임 : 하나의 학습이 끝나야 다음 학습 -> 속도 느림
